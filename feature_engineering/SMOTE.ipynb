# SMOTE - synthetic minority oversampling technique


SMOTE(Synthetic Minority Over-sampling Techique) is a techique used inn the machine learning to address imbalanced dataset where the minority class has siginficanlly fewer instance than the majority class.SMOTE involves generating synthetic instance of the minority class by interpolating between existing instance.

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,
                        weights=[0.90],random_state=12,n_redundant=0
                        )
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(X,columns = ['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
final_df = pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# Transform the dataset using SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(final_df[['f1', 'f2']], final_df['target'])
```
