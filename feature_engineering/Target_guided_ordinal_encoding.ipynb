# Target Guided Ordinal Encoding

it is a technique used to encode cateforical vairables based on their realtioship with the vairable . This encoding technique is usiful we have a catergorical variable with a large number of unique catergories and ew want to use this variable as a feature in our ml model

In Target Guided Ordinal Encoidng. we repalce each catefory in the categorical variable with a numerical value based the mean or median of the target variable for the category. This creates a monotonic relationship between the categorical variable and the target variable which can improve the predictive power of our model 

```python
import pandas as pd

df = pd.DataFrame({
    'city': ['New york','London','Paris','Tokyo','New york','Paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df[['price','city_encoded']]
```

```python
mean_bill = new_data.groupby(['time'])['total_bill'].mean().to_dict()
```
