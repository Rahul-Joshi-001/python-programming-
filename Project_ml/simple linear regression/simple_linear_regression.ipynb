# Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## read the dataset
df = pd.read_excel('weight_height.xlsx')
df.head()
```

```python
plt.scatter(df['Weight'],df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
x = df[['Weight']]
y = df['Height']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train.shape ,y_train.shape
```

```python
## standard scalar the data
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
## train our model simple linear regression model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train_scaled,y_train)
```

```python
print("the slope or coefficient of weight",linreg.coef_)
print("the intercept will be - ",linreg.intercept_)
```

```python
plt.scatter(x_train_scaled,y_train)
plt.plot(x_train_scaled,linreg.predict(x_train_scaled))
plt.show()
```
