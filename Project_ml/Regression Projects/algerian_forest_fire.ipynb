# Data Cleaning

```python
##missing values
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on REgion from 122th index, we can lmake a new col based on the Region

1 : "Bejai Region Dataset"

2 : "Sidi-Bel abbes REgion Datset"

add new col with region

```python
df.loc[:122,"Region"] = 0;
df.loc[122:,"Region"] = 1;
```

```python
df.head()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## removing the null values by dropping that row
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(122).reset_index(drop= True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in col name
df.columns =df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
## changing the required col as int datatype
df[['month','day','year','Temperature','RH','Ws']]= df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
## changing the other col to float datatype
object = [features for features in df.columns if df[features].dtype == 'O']
```

```python
for i in object:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## saving thee cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

# exploratory  data analysis

```python
## drop day month year
df_copy = df.drop({'day','month','year'},axis=1)
```

```python
df_copy.head()
```

```python
## encoding of the categoreie and classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot Density plot for all feature
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('seaborn-v0_8-dark')

# Ensure df_copy is a DataFrame with numeric columns
df_copy.hist(bins=50, figsize=(20, 15))

plt.show()
```

```python
## percentage for piechart 
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Ploting pie chart
classlabel=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabel,autopct='%1.1f%%')
plt.title("Pie chart of classes")
plt.show()
```

```python
## correaltion
df_copy.corr()
```
