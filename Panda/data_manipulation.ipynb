```python
##renaming col
df = df.rename(columns = {'Number of employees':'employes no'})
df.head(5)
```

```python
df['new_founded'] = df['Founded'].astype(float)
df.head(3)
```

```python
df['new_founded'] = df['Founded'].apply(lambda x:x*2)
df.head(3)
```

```python
##data aggregating and grouping 

grouped_mean = df.groupby('Industry')['employes no'].mean()
grouped_mean
```

```python
df.groupby(['Industry','Country'])['employes no'].sum()
```

```python
df.groupby(['Industry','Country'])['employes no'].mean()
```

```python
##aggregate multiple functions
df.groupby('Country')['employes no'].agg(['mean','sum','count'])
```

```python
## merging and joining dataframes 
df1 = pd.DataFrame({'key' : ['A','B','C'],'value1' : [1,2,3]})
df2 = pd.DataFrame({'key' : ['A','B','D'],'value1' : [4,5,6]})
```

```python
print(df1)
print("\n")
print(df2)
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
